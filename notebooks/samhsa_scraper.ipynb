```python
import pandas as pd
```

```python
df = pd.read_csv('https://buprenorphine.samhsa.gov/api/locator_export.csv')
```

```python
df1 = pd.read_csv('https://data.cityofnewyork.us/api/views/hebw-6hze/rows.csv?accessType=DOWNLOAD')
```

```python
# Change the data type of 'Zip Code' column in df1 to string
df1['Zip Code'] = df1['Zip Code'].astype(str)

# Filter df based on unique values in the 'Zip Code' column of df1
df2 = df[df['zipCode'].isin(df1['Zip Code'].unique())]
df2
```

```python
df2.to_csv('../data/samhsa_full.csv')
```
